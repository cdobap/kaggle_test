```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import precision_score, recall_score, confusion_matrix
```

```python
df = pd.read_csv("data.csv")
df = df.drop(columns=['Date'])
df = df.drop(columns=['Adj Close'])
# df = df.drop(columns=['future_mv'])
df = df.drop(columns=['Open'])
df = df.drop(columns=['High'])
df = df.drop(columns=['Low'])
df = df.drop(columns=['Close'])
# df = df.drop(columns=['Volume'])
df = df.drop(columns=['Unnamed: 0'])
```

```python
X = df.drop(columns=['future_cat'])
y = df['future_cat']

X = np.array(X)
y = np.array(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X, y = make_classification(n_features=8, random_state=0)
clf = make_pipeline(StandardScaler(), LinearSVC(random_state=0, tol=1e-5, max_iter=10000, verbose=1))
clf.fit(X_train, y_train)

predictions = clf.predict(X_test)
```

```python
print(classification_report(y_test, predictions, zero_division=0))
#print('precision score: ' + str(precision_score(y_test, predictions, average='weighted')))
#print('recall score: ' + str(recall_score(y_test, predictions, average='weighted')))
```
