```python
import pandas as pd
```

```python
df = pd.read_csv("IXIC.csv")
df = df.iloc[10000:]
```

```python
# df.loc[df['Open'] > df['Close'], 'openUp'] = 1
```

```python
li = []
for i in range(len(df)): 
    if df.iloc[i]["Open"] > df.iloc[i-1]["Close"]:
        li.append(1)
    else:
        li.append(0)
df["openUp"] = li
```

```python
li = []
for i in range(len(df)): 
    if df.iloc[i]["Open"] < df.iloc[i-1]["Close"]:
        li.append(1)
    else:
        li.append(0)
df["openDown"] = li
```

```python
li = []
for i in range(len(df)): 
    if df.iloc[i]["Open"] > df.iloc[i-1]["High"]:
        li.append(1)
    else:
        li.append(0)
df["gapUp"] = li
```

```python
li = []
for i in range(len(df)): 
    if df.iloc[i]["Open"] < df.iloc[i-1]["Low"]:
        li.append(1)
    else:
        li.append(0)
df["gapDown"] = li
```

```python
li = []
for i in range(len(df)):
    try:
        diff = (df.iloc[i+1]["Close"] - df.iloc[i]["Close"]) / df.iloc[i]["Close"]
        li.append(diff)
    except:
        li.append("None")
df["future_mv"] = li
```
